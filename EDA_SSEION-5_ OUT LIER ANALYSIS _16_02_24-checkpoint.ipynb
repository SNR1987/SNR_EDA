```python
def find_outliers(cols):
#step-1
    #step-2
    Q1=np.quantile(visa_df[cols],0.25)
    Q2=np.quantile(visa_df[cols],0.50)
    Q3=np.quantile(visa_df[cols],0.75)
    #step-3
    IQR=Q3-Q1
    UB=Q3+(1.5*IQR)
    LB=Q1-(1.5*IQR)
    UP,LB
    #step-4
    con_1=visa_df[cols]>UB
    con_2=visa_df[cols]<LB
    res_outliers=visa_df[con_1 | con_2]
    
    print(f"{cols} has {len(res_outliers)} outliers ")
    #print(res_outliers)
num_cols=visa_df.select_dtypes(exclude="object").columns
for cols in num_cols:
    find_outliers(cols)
```

```python
#step-1
p_wage=visa_df["prevailing_wage"]
#step-2
Q1=np.quantile(p_wage,0.25)
Q2=np.quantile(p_wage,0.50)
Q3=np.quantile(p_wage,0.75)
#step-3
IQR=Q3-Q1
UB=Q3+(1.5*IQR)
LB=Q1-(1.5*IQR)
UP,LB
#step-4
con_1=visa_df["prevailing_wage"]>UB
con_2=visa_df["prevailing_wage"]<LB
res=visa_df[con_1 | con_2]
res
#non out liers data 
con_11=visa_df["prevailing_wage"]<UB
con_22=visa_df["prevailing_wage"]>LB
res_non_outliers=visa_df[con_11 & con_22]
res_non_outliers
```
