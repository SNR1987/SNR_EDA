```python
continent_data=dict(visa_df['continent'].value_counts())
```

```python
keys=continent_data.keys()
values=continent_data.values()
keys,values
```

```python
continent_df=pd.DataFrame(zip(keys,values),columns=["continent","count"])
continent_df.to_csv("continent_df.csv",index=False)
print(continent_df)
```

```python
# bar-chart
# bar chart required one catigorical column ad one numarical column
# continent df consider as frequency table
# continent column on x =axis and count column on y axis .
plt.figure(figsize=(10,5))
plt.bar('continent','count',data=continent_df)
plt.xlabel("Continent")
plt.ylabel("Count")
plt.title("Continent bar chat")
plt.savefig("Continent-count bar chart.png")
plt.show()
```

```python
# bar chart required 2 things.
# x- axis is the catigorical column
# y- axis requires numarical column
```
